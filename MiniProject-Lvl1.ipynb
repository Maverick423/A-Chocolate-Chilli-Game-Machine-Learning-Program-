### Rules

#### What it does
Prints all the rules of the game on the screen.
#### Implementation (how it works)
Since all i had to do was output the rules for the i used System.out.println() to print out all the rules on the screen. There are 7 rules printed out along with the title, Chocolate Chilli Game Rules.

````java
public static void Rules()
{
    System.out.println("Chocolate Chilli Game Rules");
    System.out.println("");
    System.out.println("1. There is a pile of chocolates on the table.");
    System.out.println("2. Two players take turns");
    System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
    System.out.println("4. Players must take at least one chocolate each turn.");
    System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
    System.out.println("6. The game ends as soon as there are no chocolates left.");
    System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.");
    return;
}
````

### Introduction

#### What it does
Asks user their name and how many chocolates they wish to start with. Than prints Thanks, (name)! There are (Chocolates) chocolates on the table. I will go first.

#### Implementation (how it works)   
To get inputs from user i had to import java.util library to use the Scanner class. Than i initialised the class. Made 2 variables for name and number of chocolates to start with which stored the inputs from user.

````java
import java.util.Scanner;
public static int Introduction()
{
    Scanner scanner = new Scanner(System.in);
    String name;
    int numChocolates;
    System.out.println("What's your name?");
    name = scanner.nextLine();
    System.out.println("How many chocolates they wish to start with?");
    numChocolates = Integer.parseInt(scanner.nextLine());
    System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.");
    return numChocolates;
}
````

## The complete program
This version will only compile here. To run it copy it into a file called initials.java on your local computer and compile and run it there.

````java
// Maverick Anushk Fernandes
// 23/10/2025
// VERSION 1

/* States the rules of the game. Than asks the user to input thier name and how many chocolates they wish to start with.
At the end of the method Introduction it returns the number of chocolate whihc is stored in numChocolate in main function.*/

import java.util.Scanner; // Needed to make Scanner available

class ChocolateChilliGame
{ // START ChocolateChilliGame
    public static void main (String [] arg)
    { // START main
        int numChocolate; // number of chocolate to start with
        Rules(); // Method call
        numChocolate = Introduction(); // Method call
        return;
    } // END main
    
    public static void Rules()
    { // START Rules
        System.out.println("Chocolate Chilli Game Rules");
        System.out.println(""); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.");
        System.out.println("");
        return;
    } // END Rules

    public static int Introduction()
    { // START Introduction
        Scanner scanner = new Scanner(System.in);
        String name; // users name
        int numChocolates; // number of chocolate to start with
        System.out.println("What's your name?");
        name = scanner.nextLine();
        System.out.println("How many chocolates they wish to start with?");
        numChocolates = Integer.parseInt(scanner.nextLine()); // convert string to int
        System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.");
        return numChocolates;
    } // END Introduction
    
} // END ChocolateChilliGame
````
