#### Summary
As above, but the program now takes chocolates from the table for its turn (say 3 every time it is run), remembering the initial value but also keeping track of the current number left (so uses 2 variables for this). The main method is structured as a sequence of calls to methods: 1) call a method to describe the rules 2) call a method to ask for the number of chocolates (this method returns their answer) 3) repeat their answer “There are 11 chocolates …” 4) Call a method to say the move (take 3). 

````java
import java.util.Scanner; // Needed to make Scanner available

class ChocolateChilliGame
{ // START ChocolateChilliGame
    public static void main (String [] arg)
    { // START main
        Scanner scanner = new Scanner(System.in);
        int numChocolate; // number of chocolate to start with
        Rules(); // Method call
        numChocolate = Introduction(scanner); // Method call
        GameLoop(scanner, numChocolate);
        return;
    } // END main
    
    public static void Rules()
    { // START Rules
        System.out.println("Chocolate Chilli Game Rules\n"); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.\n");
        return;
    } // END Rules

    public static int Introduction(Scanner scanner)
    { // START Introduction
        String name; // users name
        int numChocolates; // number of chocolate to start with
        System.out.println("What's your name?");
        name = scanner.nextLine();
        System.out.println("How many chocolates they wish to start with?");
        numChocolates = Integer.parseInt(scanner.nextLine()); // convert string to int
        System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.\n");
        return numChocolates;
    } // END Introduction
    
    public static void GameLoop(Scanner scanner, int initialChocolates)
    {
        int leftChocolates = initialChocolates; // set intitial number of chocolates to number of chocolates left
        int pickedChocolates;
        while (leftChocolates > 0) // runs until there is no more chocolates left
        {
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
            if (pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3) // checks if the user entered vaild inputs
            {
                leftChocolates -= pickedChocolates; // subtracts by input
            } else 
            {
                continue; // carrys on while next loop cycle
            }
            System.out.println("There are " + leftChocolates + " left.\n");
        }
        return;
    }
} // END ChocolateChilliGame
````

### Player

#### What it does
Takes the the total number of chocolates they start with as a parameter. Asks user how many chocolates they would like to eat. It keeps on asking until a suitable answer is given (1 to 3). Than it subtracts the amount user has eaten by the total amount. It returns the number of remaining chocolates left.

#### Implementation (how it works)
I used the same code from the GameLoop the only difference is that i got rid of the if statement and while loop and replaced it with a do  while loop. The reason i chose do while is because i want the user to input a number first than check if the number inputted is within range. If its not than it repeats until right input is given.


````java
public static int Player(Scanner scanner, int chocolatesLeft)
    { // START Player
        int pickedChocolates;
        do { // do while loop so it runs once before checking condition
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
        } while (!(pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3)); // makes sure user pick 1, 2 or 3
        chocolatesLeft -= pickedChocolates; // subtracts by input
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } //END Player
````

### Machine

#### What it does
Takes the the total number of chocolates they start with as a parameter. Uses random library to pick a number between 1 and 3, than prints that number out telling the user what it picked. Than it subtracts the amount user has eaten by the total amount. It returns the number of remaining chocolates left.

#### Implementation (how it works)
There are 2 parameter random class and total number of chocolates left. Using the library i was able to get a random number between 1 and 3. I had to use random.nextInt(3)+1 because random.nextInt(1,3) doesn't work with older java. After picking number it subtracts the amount machine has eaten by the total amount. It returns the number of remaining chocolates left.

````java
public static int Machine(Random random, int chocolatesLeft) 
    { //START Machine
        int pickedChocolates = random.nextInt(3)+1; // (3) pickes 3 numbers from 0 so (0, 1, 2) hence +1
        System.out.println("The machine has picked " + pickedChocolates);
        chocolatesLeft -= pickedChocolates; // subtracts by input
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } // END Machine
````

### End

#### What it does
Checks if the game has ended. If the game has ended than returns true, if its still running that false.
#### Implementation (how it works)
Added an if statement that checks if the number of chocolates if lower than 0 meaning not more chocolates left and if there aren't any chocolates left than return true, meaning the game has ended. If there are chocolates left than returns, true game carries on. 

````java
public static boolean End(int chocolatesLeft)
{ // START End
    if (chocolatesLeft<=0) // checks if game has ended
    {
        return true; // ended
    }else
    {
        return false; // not ended
    }
} // END End
````

## The complete program
This version will only compile here. To run it copy it into a file called initials.java on your local computer and compile and run it there.

````java
// Maverick Anushk Fernandes
// 23/10/2025
// VERSION 1

/* 
States the rules of the game. Than asks the user to input thier name and how many chocolates they wish to start with.
At the end of the method Introduction it returns the number of chocolate which is stored in numChocolate in main function.
Machine and user take turns to eat the chocolates until there are 0 or less than 0 chocolates left.
*/

import java.util.Scanner; // Needed to make Scanner available
import java.util.Random; // Needed to make Random available

class ChocolateChilliGame
{ // START ChocolateChilliGame
    public static void main (String [] arg)
    { // START main
        Random random = new Random();
        Scanner scanner = new Scanner(System.in);
        int numChocolate; // number of chocolate to start with
        Rules(); // Method call
        numChocolate = Introduction(scanner); // Method call
        GameLoop(scanner, random, numChocolate);
        return;
    } // END main
    
    public static void Rules()
    { // START Rules
        System.out.println("Chocolate Chilli Game Rules\n"); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.\n");
        return;
    } // END Rules

    public static int Introduction(Scanner scanner)
    { // START Introduction
        String name; // users name
        int numChocolates; // number of chocolate to start with
        System.out.println("What's your name?");
        name = scanner.nextLine();
        System.out.println("How many chocolates they wish to start with?");
        numChocolates = Integer.parseInt(scanner.nextLine()); // convert string to int
        System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.\n");
        return numChocolates;
    } // END Introduction
    
    public static void GameLoop(Scanner scanner, Random random, int initialChocolates)
    { // START GameLoop
        int leftChocolates = initialChocolates; // set intitial number of chocolates to number of chocolates left
        while (!(End(leftChocolates))) // runs until there is no more chocolates left
            {
                leftChocolates = Machine(random, leftChocolates); // machines turn
                leftChocolates = Player(scanner, leftChocolates); // players turn
            }
        return;
    } // END GameLoop

    public static int Player(Scanner scanner, int chocolatesLeft)
    { // START Player
        int pickedChocolates;
        do { // do while loop so it runs once before checking condition
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
        } while (!(pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3)); // makes sure user pick 1, 2 or 3
        chocolatesLeft -= pickedChocolates; // subtracts by input
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } //END Player

    public static int Machine(Random random, int chocolatesLeft) 
    { //START Machine
        int pickedChocolates = random.nextInt(3)+1; // (3) pickes 3 numbers from 0 so (0, 1, 2) hence +1
        System.out.println("The machine has picked " + pickedChocolates);
        chocolatesLeft -= pickedChocolates; // subtracts by input
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } // END Machine

    public static boolean End(int chocolatesLeft)
    { // START End
        if (chocolatesLeft<=0) // checks if game has ended
        {
            return true; // ended
        }else
        {
            return false; // not ended
        }
    } // END End

} // END ChocolateChilliGame
````
