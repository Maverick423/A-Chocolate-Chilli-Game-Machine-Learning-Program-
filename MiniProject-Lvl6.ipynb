## Summary of the Question 
As above but the program also keeps an array of records. Each
array entry corresponds to a number of chocolates left so possible state of the game. Each record has three
boolean fields standing for “eat one chocolate”, “eat two chocolates” or “eat three chocolates’. All are set to
true initially except for at array position one in which only the value for “eat one ..” is true; and array
position two in which only “eat one…” and “eat two …” are true. The program then only makes moves on its
turn corresponding to the record being true for that move. Array position 0 is not used. If the program makes
a move and then immediately loses then the entry for that move is set to false. The program now resigns
immediately if all booleans in the record for the current position are false (as all lead to losses). It then
changes the boolean flag for its last move (to false so it won’t make that move again if in the same position).

### MemoryRemove

Than it checks if the inner memory is all false by call the method CheckMemory and if it is than calls EditMemory else nothing happens.

There are 4 parameters memory, moves, chocolates left and chocolate picked.
````java
public static void MemoryRemove(ArrayList<ArrayList<Boolean>> memory, ArrayList<Integer> moves,  int chocolatesLeft, int chocolatePicked) { // START MemoryRemove
    memory.get(chocolatesLeft-1).set(chocolatePicked-1,false); // since 2d gets inner list [[],[],[]] from it makes whatever was picked false 
    if (CheckMemory(memory,chocolatesLeft).isEmpty()){ // checks if all of it is false
        EditMemory(memory,moves); // changes the previous move 
    }
    return;
} // END MemoryRemove
````

### CheckMemory

#### What it does
Checks in memory which of the values are true so if [true, false, true] than in a new array it will have [1,3] after the loop goes through the inner loop. Than it will return the new array which has the number of chocolates are available to eat.
#### Implementation (how it works)
listNum is a dynamic array that will hold the number of chocolates are available to eat. I used a for loop that goes through the inner array and checks if each one is true and if it is true than it will be added to the listNum after adding one to it. This is because rather than 0 to 2 i want it to go 1 to 3 so its easier to pick the chocolate machine wants to eat.

````java
public static ArrayList<Integer> CheckMemory(ArrayList<ArrayList<Boolean>> memory, int chocolatesLeft){
    ArrayList<Integer> listNum = new ArrayList<>(); // dynamic 1d array
    for (int i = 0; i<memory.get(chocolatesLeft-1).size(); i++){ // [[],[]] gets the size of the inner array 
        if (memory.get(chocolatesLeft-1).get(i)){ // checks if true eg [[true,false,true],[]]
            listNum.add(i+1); // if true it will return the index value so new list will be [1,3] from above
        }
    }
    return listNum;
}
````

### EditMemory

#### What it does
Checks if the size of the moves array which stores all the moves done by player and machine. This method is used when the inner array has all false [false,false,false] than it will go to the one before that move and make it false. For example if 4 is all false and before that was 10 [true, true,true] machine picked 3 player picked 3 now its 4 left which is all false so it update 10 to [true,true,false].
#### Implementation (how it works)
The if statement checks if the size of array is less than 4, for validation purpose if i didn't have that than it would cause errors in the code. The for loop goes through the moves array and checks which one is all false in the memory. When it finds the one which is empty it will first work out the number picked by machine by subtracting current machines move and past machine move. Than it goes to that previous machine choice and makes the number it picked false.

````java
public static void EditMemory(ArrayList<ArrayList<Boolean>> memory, ArrayList<Integer> moves) { // END EditMemory
    if (moves == null || moves.size() < 4) // validation 
        return; 
    for (int i = moves.size()-2; i >= 0; i-=2){ // finds the position of the empty (all false) inner array
        if (CheckMemory(memory, moves.get(i)).isEmpty()){ // checks if there are possible moves list is empty
            int pickedChocolate = moves.get(i)-moves.get(i-2); // works out what machine picked 
            memory.get(moves.get(i)-1).set(pickedChocolate-1,false); // [12,8,5,4] is all is false for 4 in the inner than will go to 8 and change that  
        }
    }
    return;
} // END EditMemory 
````

````java
// Maverick Anushk Fernandes
// 24/10/2025
// VERSION 1

/* 
States the rules of the game. Than asks the user to input thier name and how many chocolates they wish to start with.
At the end of the method Introduction it returns the number of chocolate which is stored in numChocolate in main function.
Machine and user take turns to eat the chocolates until there are 0 left for a certain amount of games which the user decides.
At the end depending on who won the most it will say "Winner is machine" or "Winner is (name of user)" or "Its a draw".
Stores the record in memory, finds bad moves and avoids them.
*/

import java.util.Scanner; // Needed to make Scanner available
import java.util.Random; // Needed to make Random available
import java.util.ArrayList; // Needed to make ArrayList available
import java.util.Arrays; // Needed to make Arrays available

class ChocolateChilliGame { // START ChocolateChilliGame
    
    public static void main (String [] arg) { // START main
        Random random = new Random();
        Scanner scanner = new Scanner(System.in);
        String[] playerInfo = new String[3]; // number of chocolate to start with
        Rules(); // Method call
        playerInfo = Introduction(scanner); // Method call
        GameLoop(scanner, random, playerInfo);
        scanner.close();
        return;
    } // END main
    
    public static void Rules() { // START Rules
        System.out.println("Chocolate Chilli Game Rules\n"); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.\n");
        return;
    } // END Rules

    public static String[] Introduction(Scanner scanner) { // START Introduction
        String[] intro = new String[3]; // [name, intial chocolate, total games]
        System.out.println("What's your name?");
        intro[0] = scanner.nextLine(); // name
        do { // validation
        System.out.println("How many chocolates would you like to start with? (Greater than 10 inclusive)");
        intro[1] = scanner.nextLine(); // intial chocolate
        }while (Integer.parseInt(intro[1])<10);
        do { // validation
        System.out.println("How many games do you want to play?");
        intro[2] = scanner.nextLine(); // total games
        }while(Integer.parseInt(intro[2])<=0);
        System.out.println("Thanks, " + intro[0] + "! There are " + intro[1] + " chocolates on the table. I will go first.\n");
        return intro;
    } // END Introduction
    
    public static void GameLoop(Scanner scanner, Random random, String[] playerInfo) { // START GameLoop
        int leftChocolates = Integer.parseInt(playerInfo[1]); // initial chocolate
        int totalGame = Integer.parseInt(playerInfo[2]); // number of games to play
        boolean machineWon;
        String winner = null;
        int[] allGame = new int[totalGame];
        ArrayList<Integer> listNum = new ArrayList<>();
        ArrayList<ArrayList<Boolean>> memory = Memory();
        for (int i = 1; i<=totalGame; i++){
                System.out.println("Game " + i);
                do{ // so all variables are initialised 
                    leftChocolates = Machine(memory, random, leftChocolates); // machines turn
                    machineWon = true;
                    listNum.add(leftChocolates);
                    if (leftChocolates<=0) { // checks if there are any chocolates left 
                        break; // breaks do while loop
                    }
                    leftChocolates = Player(scanner, leftChocolates); // players turn
                    machineWon = false;
                    listNum.add(leftChocolates);
                } while (!(leftChocolates<=0)); // runs until there is no more chocolates left
            leftChocolates = Integer.parseInt(playerInfo[1]); // gets original starting number of chocolates as new round
            allGame = Score(playerInfo,machineWon,allGame); //records who won
            winner = Result(playerInfo,allGame);
            }
        if (winner.equals("Draw")){ // draw
            System.out.println("Its a draw");
        }else{
            System.out.println("Winner is " + winner); // someone won
        }
        return;
    } // END GameLoop

    public static int Player(Scanner scanner, int chocolatesLeft) { // START Player
        int pickedChocolates;
        do { // do while loop so it runs once before checking condition
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
        } while (!(pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3)); // makes sure user pick 1, 2 or 3
        if (pickedChocolates > chocolatesLeft) {
            System.out.println("There are only " + chocolatesLeft + " left. Try again\n");
            chocolatesLeft = Player(scanner, chocolatesLeft);
        }else {
            chocolatesLeft -= pickedChocolates; // subtracts by input
        }
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } //END Player
    
    public static int Machine(ArrayList<ArrayList<Boolean>> memory, Random random, int chocolatesLeft) { //START Machine
        int pickedChocolates = MemoryGetNum(random, memory, chocolatesLeft);
        ArrayList<Integer> moveSet = new ArrayList<Integer>();
        ArrayList<Integer> listNum = CheckMemory(memory, chocolatesLeft); // list of all moves
        if (listNum.isEmpty()) { 
            return 0; // all move set is false so can't pick 
        } 
        System.out.println("The machine has picked " + pickedChocolates);
        if (pickedChocolates > chocolatesLeft) {
            MemoryRemove(memory, null, chocolatesLeft, pickedChocolates); // if what machine picked exceeds number of chocolate left than removes it from move set
            chocolatesLeft = Machine(memory, random, chocolatesLeft); // calls it again to get different value
        }else {
            chocolatesLeft -= pickedChocolates; // subtracts by input
        }
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left       
        return chocolatesLeft;
    } // END Machine
    
    public static int[] Score(String[] infoPlayer, boolean machineWin, int[] gameTotal) { // START Score
        int j = gameTotal.length; // j is max
        for(int i = 0; i<gameTotal.length; i++) {
            if (gameTotal[i] == 0) { // when initialised contains all 0 
                j-=1; // gets the position of the next empty space
            }
        }
        if (machineWin) { // checks boolean value
            gameTotal[j] = 1; // machine won
        }else {
            gameTotal[j] = 2; // player won
        }
        return gameTotal;
    } // END Score
    
    public static String Result(String[] infoPlayer, int[] gameResult) { // START Result
        String name = infoPlayer[0]; //S first value has name
        int playerScore = 0; // start score
        int machineScore = 0; // start score
        for (int i = 0; i<gameResult.length; i++) {
            if (gameResult[i] == 2) { // player won 
                playerScore += 1;
            }else if (gameResult[i] == 1) { // machine won
                machineScore += 1;
            }
        }
        System.out.println(name + ": " + playerScore); // print score 
        System.out.println("Machine: " + machineScore + "\n"); // print score
        if (playerScore>machineScore) { // player won
            return infoPlayer[0];
        }else if (playerScore<machineScore) { // machine won
            return "Machine";
        }else {
            return "Draw"; // draw
        }
    } // END Result

    public static ArrayList<ArrayList<Boolean>> Memory() { // START Memory
        ArrayList<ArrayList<Boolean>> memory = new ArrayList<>(); // 2d array with dynamic memory 
        for (int i = 0; i < 100; i++){
            memory.add(new ArrayList<>(Arrays.asList(true,true,true))); // initialising the dynamic array with 3 numbers
        }
        return memory;
    } // END Memory

    public static void MemoryRemove(ArrayList<ArrayList<Boolean>> memory, ArrayList<Integer> moves,  int chocolatesLeft, int chocolatePicked) { // START MemoryRemove
        memory.get(chocolatesLeft-1).set(chocolatePicked-1,false); // since 2d gets inner list [[],[],[]] from it makes whatever was picked false 
        if (CheckMemory(memory,chocolatesLeft).isEmpty()){ // checks if all of it is false
            EditMemory(memory,moves); // changes the previous move 
        }
        return;
    } // END MemoryRemove
    
    public static int MemoryGetNum(Random random, ArrayList<ArrayList<Boolean>> memory, int chocolatesLeft) { // START MemoryGetNum
        ArrayList<Integer> listNum = new ArrayList<>(); // dynamic 1d array
        listNum = CheckMemory(memory, chocolatesLeft); // gets all possible move so (1,2,3) based on which is true 
        int pickedChocolates = listNum.get(random.nextInt(listNum.size())); // randomly picks one of the availabe options
        return pickedChocolates;
    } // END MemoryGetNum

    public static ArrayList<Integer> CheckMemory(ArrayList<ArrayList<Boolean>> memory, int chocolatesLeft){
        ArrayList<Integer> listNum = new ArrayList<>(); // dynamic 1d array
        for (int i = 0; i<memory.get(chocolatesLeft-1).size(); i++){ // [[],[]] gets the size of the inner array 
            if (memory.get(chocolatesLeft-1).get(i)){ // checks if true eg [[true,false,true],[]]
                listNum.add(i+1); // if true it will return the index value so new list will be [1,3] from above
            }
        }
        return listNum;
    }

    public static void EditMemory(ArrayList<ArrayList<Boolean>> memory, ArrayList<Integer> moves) { // END EditMemory
        if (moves == null || moves.size() < 4) // validation 
            return;
        for (int i = moves.size()-2; i >= 0; i-=2){ // finds the position of the empty (all false) inner array
            if (CheckMemory(memory, moves.get(i)).isEmpty()){ // checks if there are possible moves list is empty
                int pickedChocolate = moves.get(i)-moves.get(i-2); // works out what machine picked 
                memory.get(moves.get(i)-1).set(pickedChocolate-1,false); // [12,8,5,4] is all is false for 4 in the inner than will go to 8 and change that  
            }
        }
        return;
    } // END EditMemory 

   
} // END ChocolateChilliGame
````
