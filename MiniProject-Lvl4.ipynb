````java
import java.util.Scanner; // Needed to make Scanner available
import java.util.Random; // Needed to make Random available

class ChocolateChilliGame
{ // START ChocolateChilliGame
    public static void main (String [] arg)
    { // START main
        Random random = new Random();
        Scanner scanner = new Scanner(System.in);
        int numChocolate; // number of chocolate to start with
        Rules(); // Method call
        numChocolate = Introduction(scanner); // Method call
        GameLoop(scanner, random, numChocolate);
        return;
    } // END main
    
    public static void Rules()
    { // START Rules
        System.out.println("Chocolate Chilli Game Rules\n"); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.\n");
        return;
    } // END Rules

    public static int Introduction(Scanner scanner)
    { // START Introduction
        String name; // users name
        int numChocolates; // number of chocolate to start with
        System.out.println("What's your name?");
        name = scanner.nextLine();
        System.out.println("How many chocolates they wish to start with?");
        numChocolates = Integer.parseInt(scanner.nextLine()); // convert string to int
        System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.\n");
        return numChocolates;
    } // END Introduction
    
    public static void GameLoop(Scanner scanner, Random random, int initialChocolates)
    { // START GameLoop
        int leftChocolates = initialChocolates; // set intitial number of chocolates to number of chocolates left
        while (!(End(leftChocolates))) // runs until there is no more chocolates left
            {
                leftChocolates = Machine(random, leftChocolates); // machines turn
                leftChocolates = Player(scanner, leftChocolates); // players turn
            }
        return;
    } // END GameLoop

    public static int Player(Scanner scanner, int chocolatesLeft)
    { // START Player
        int pickedChocolates;
        do { // do while loop so it runs once before checking condition
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
        } while (!(pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3)); // makes sure user pick 1, 2 or 3
        chocolatesLeft -= pickedChocolates; // subtracts by input
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } //END Player

    public static int Machine(Random random, int chocolatesLeft) 
    { //START Machine
        int pickedChocolates = random.nextInt(3)+1; // (3) pickes 3 numbers from 0 so (0, 1, 2) hence +1
        System.out.println("The machine has picked " + pickedChocolates);
        chocolatesLeft -= pickedChocolates; // subtracts by input
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } // END Machine

    public static boolean End(int chocolatesLeft)
    { // START End
        if (chocolatesLeft<=0) // checks if game has ended
        {
            return true; // ended
        }else
        {
            return false; // not ended
        }
    } // END End

} // END ChocolateChilliGame
````

### GameLoop (updated)

#### What it does
Calls the End method and checks its the end of game if its not end that the while loop loops until end is reached. In the loop it calls Machine method which is machines turn playing than checks if there are any chocolate left and if there are than goes to players turn.
#### Implementation (how it works)
!(End(leftChocolates, false) calls the method if it returns false (meaning not end) not false is true making the while loop run. In the loop it starts on machines turn and goes to players turn after it does it until there are no more chocolates left.

````java
public static void GameLoop(Scanner scanner, Random random, int initialChocolates)
{ // START GameLoop
    int leftChocolates = initialChocolates; // set intitial number of chocolates to number of chocolates left
    while (!(End(leftChocolates, false))) // runs until there is no more chocolates left
        {
            leftChocolates = Machine(random, leftChocolates); // machines turn
            if (End(leftChocolates, true))
            {
                break;
            }
            leftChocolates = Player(scanner, leftChocolates); // players turn
        }
    return;
} // END GameLoop
````

````java
// Maverick Anushk Fernandes
// 24/10/2025
// VERSION 1

/* 
States the rules of the game. Than asks the user to input thier name and how many chocolates they wish to start with.
At the end of the method Introduction it returns the number of chocolate which is stored in numChocolate in main function.
Machine and user take turns to eat the chocolates until there are 0 left. Depending on who wins it will say "machine wins" or "you win".
*/

import java.util.Scanner; // Needed to make Scanner available
import java.util.Random; // Needed to make Random available

class ChocolateChilliGame
{ // START ChocolateChilliGame
    public static void main (String [] arg)
    { // START main
        Random random = new Random();
        Scanner scanner = new Scanner(System.in);
        int numChocolate; // number of chocolate to start with
        Rules(); // Method call
        numChocolate = Introduction(scanner); // Method call
        GameLoop(scanner, random, numChocolate);
        return;
    } // END main
    
    public static void Rules()
    { // START Rules
        System.out.println("Chocolate Chilli Game Rules\n"); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.\n");
        return;
    } // END Rules

    public static int Introduction(Scanner scanner)
    { // START Introduction
        String name; // users name
        int numChocolates; // number of chocolate to start with
        System.out.println("What's your name?");
        name = scanner.nextLine();
        System.out.println("How many chocolates they wish to start with?");
        numChocolates = Integer.parseInt(scanner.nextLine()); // convert string to int
        System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.\n");
        return numChocolates;
    } // END Introduction
    
    public static void GameLoop(Scanner scanner, Random random, int initialChocolates)
    { // START GameLoop
        int leftChocolates = initialChocolates; // set intitial number of chocolates to number of chocolates left
        while (!(End(leftChocolates, false))) // runs until there is no more chocolates left
            {
                leftChocolates = Machine(random, leftChocolates); // machines turn
                if (End(leftChocolates, true))
                {
                    break;
                }
                leftChocolates = Player(scanner, leftChocolates); // players turn
            }
        return;
    } // END GameLoop

    public static int Player(Scanner scanner, int chocolatesLeft)
    { // START Player
        int pickedChocolates;
        do { // do while loop so it runs once before checking condition
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
        } while (!(pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3)); // makes sure user pick 1, 2 or 3
        chocolatesLeft -= pickedChocolates; // subtracts by input
        if (chocolatesLeft<0)
        {
            chocolatesLeft = 0;
        }
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } //END Player
    
    public static int Machine(Random random, int chocolatesLeft) 
    { //START Machine
        final int pickMax=3;
        int pickedChocolates = random.nextInt(pickMax)+1; // (pickMax) pickes 3 numbers from 0 so (0, 1, 2) hence +1
        System.out.println("The machine has picked " + pickedChocolates);
        chocolatesLeft -= pickedChocolates; // subtracts by input
        if (chocolatesLeft<0)
        {
            chocolatesLeft = 0;
        }
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } // END Machine
    
    public static boolean End(int chocolatesLeft, boolean winner)
    { // START End
        if (chocolatesLeft<=0) // checks if game has ended
        {
            if (winner)
            {
                System.out.println("Machine wins"); // machine win
            }else
            {
                System.out.println("You win"); // machine lose
            }
            return true; // ended
        }else
        {
            return false; // not ended
        }
    } // END End

} // END ChocolateChilliGame
````
