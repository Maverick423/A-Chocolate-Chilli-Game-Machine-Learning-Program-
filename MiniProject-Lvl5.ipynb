#### Summary
As above, but now players take it in turns to eat chocolate(s) repeatedly until it is the end of the game (no chocolates left). A separate method is called for each player’s turn. The new total is returned and passed to the check end method each time. The program also checks a move doesn’t take the number of chocolates below 0.

````java
import java.util.Scanner; // Needed to make Scanner available
import java.util.Random; // Needed to make Random available

class ChocolateChilliGame
{ // START ChocolateChilliGame
    public static void main (String [] arg)
    { // START main
        Random random = new Random();
        Scanner scanner = new Scanner(System.in);
        int numChocolate; // number of chocolate to start with
        Rules(); // Method call
        numChocolate = Introduction(scanner); // Method call
        GameLoop(scanner, random, numChocolate);
        return;
    } // END main
    
    public static void Rules()
    { // START Rules
        System.out.println("Chocolate Chilli Game Rules\n"); // Rules
        System.out.println("1. There is a pile of chocolates on the table.");
        System.out.println("2. Two players take turns");
        System.out.println("3. On your turn, you must eat 1, 2, or 3 chocolates from the pile.");
        System.out.println("4. Players must take at least one chocolate each turn.");
        System.out.println("5. The player who has to take the last chocolate is the loser - they have to eat the chilli instead of chocolate!");
        System.out.println("6. The game ends as soon as there are no chocolates left.");
        System.out.println("7. The aim is to avoid taking the last chocolate - so you must try to force your opponent to take it.\n");
        return;
    } // END Rules

    public static int Introduction(Scanner scanner)
    { // START Introduction
        String name; // users name
        int numChocolates; // number of chocolate to start with
        System.out.println("What's your name?");
        name = scanner.nextLine();
        System.out.println("How many chocolates they wish to start with?");
        numChocolates = Integer.parseInt(scanner.nextLine()); // convert string to int
        System.out.println("Thanks, " + name + "! There are " + numChocolates + " chocolates on the table. I will go first.\n");
        return numChocolates;
    } // END Introduction
    
    public static void GameLoop(Scanner scanner, Random random, int initialChocolates)
    { // START GameLoop
        int leftChocolates = initialChocolates; // set intitial number of chocolates to number of chocolates left
        while (!(End(leftChocolates, false))) // runs until there is no more chocolates left
            {
                leftChocolates = Machine(random, leftChocolates); // machines turn
                if (End(leftChocolates, true))
                {
                    break;
                }
                leftChocolates = Player(scanner, leftChocolates); // players turn
            }
        return;
    } // END GameLoop

    public static int Player(Scanner scanner, int chocolatesLeft)
    { // START Player
        int pickedChocolates;
        do { // do while loop so it runs once before checking condition
            System.out.println("How many chocolates do you wish to eat? (1, 2 or 3)");
            pickedChocolates = Integer.parseInt(scanner.nextLine()); // converts string to int
        } while (!(pickedChocolates == 1 || pickedChocolates == 2 || pickedChocolates == 3)); // makes sure user pick 1, 2 or 3
        chocolatesLeft -= pickedChocolates; // subtracts by input
        if (chocolatesLeft<0)
        {
            chocolatesLeft = 0;
        }
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } //END Player
    
    public static int Machine(Random random, int chocolatesLeft) 
    { //START Machine
        final int pickMax=3;
        int pickedChocolates = random.nextInt(pickMax)+1; // (pickMax) pickes 3 numbers from 0 so (0, 1, 2) hence +1
        System.out.println("The machine has picked " + pickedChocolates);
        chocolatesLeft -= pickedChocolates; // subtracts by input
        if (chocolatesLeft<0)
        {
            chocolatesLeft = 0;
        }
        System.out.println("There are " + chocolatesLeft + " left.\n"); // total number of chocolates left
        return chocolatesLeft;
    } // END Machine
    
    public static boolean End(int chocolatesLeft, boolean winner)
    { // START End
        if (chocolatesLeft<=0) // checks if game has ended
        {
            if (winner)
            {
                System.out.println("Machine wins"); // machine win
            }else
            {
                System.out.println("You win"); // machine lose
            }
            return true; // ended
        }else
        {
            return false; // not ended
        }
    } // END End

} // END ChocolateChilliGame
````

### Score

#### What it does
Takes in 3 parameter 2 of which are array and one is boolean. It goes through the array and gets the position of the next available space. Than it checks who won using the boolean variable and add that point in the gameTotal array.

#### Implementation (how it works)
Has 3 parameter infoPlayer has name, number to play to, and total games. machineWin has true or false so if machine won than true. And lastly gameTotal has the points of who won in total. First j is assigned the total length of array and if there are empty space (0) than subtracts j from that many times. Than stores who won machine or player in that position and returns the array at the end.

````java
public static int[] Score(String[] infoPlayer, boolean machineWin, int[] gameTotal)
{ // START Score
    int j = gameTotal.length; // j is max
    for(int i = 0; i<gameTotal.length; i++)
    {
        if (gameTotal[i] == 0) // when initialised contains all 0
        {
            j-=1; // gets the position of the next empty space
        }
    }
    if (machineWin) // checks boolean value
    {
        gameTotal[j] = 1; // machine won
    }else 
    {
        gameTotal[j] = 2; // player won
    }
    return gameTotal;
} // END Score
````

### Result

#### What it does
Takes in 2 parameter which are array. The infoPlayer has the name number of games etc. gameResult has the record of who won every round. Using the first array to find the total number of games, it goes through the entire array and checks who won. At the end it says the score of the round.
#### Implementation (how it works)
Takes in 2 parameter which are array. The infoPlayer has the name number of games etc. gameResult has the record of who won every round. Using the first array to find the total number of games, it goes through the entire array and checks who won. This is done by checking if the value in the array is 1 (machine won) or 2 (player won). Adds the total to the variable than prints out who won. And returns what happened, who won, machine or player or draw. 

````java
public static String Result(String[] infoPlayer, int[] gameResult)
{ // START Result
    String name = infoPlayer[0]; // first value has name
    int playerScore = 0; // start score
    int machineScore = 0; // start score
    for (int i = 0; i<gameResult.length; i++)
    {
        if (gameResult[i] == 2) // player won
        {
            playerScore += 1;
        }else if (gameResult[i] == 1) // machine won
        {
            machineScore += 1;
        }
    }
    System.out.println(name + ": " + playerScore); // print score 
    System.out.println("Machine: " + machineScore + "\n"); // print score
    if (playerScore>machineScore) // player won
    {
        return infoPlayer[0];
    }else if (playerScore<machineScore) // machine won
    {
        return "Machine";
    }else
    {
        return "Draw"; // draw
    }
} // END Result
````

### GameLoop (updated)

#### What it does
The method does all the rounds in the game. The for loop is the amount of rounds there are inside it the player and machine method is called repeatedly for them to choose how many chocolates they want to eat until there are no more chocolates. After all the rounds it will output the score and the winner.  
#### Implementation (how it works)
This is the game loop where all the game handling happens. I changed the while loop to do while loop so that it runs at least once. In the loop it starts with machines turn calling the machine function that plays its turn than checks if there are any chocolates than goes to players turn, if there aren't than ends the loop. Than calls Score method to record who won and after that calls Result method to get the score. After all the games are done it will print who won.

````java
public static void GameLoop(Scanner scanner, Random random, String[] playerInfo)
{ // START GameLoop
    int leftChocolates = Integer.parseInt(playerInfo[1]); // initial chocolate
    int totalGame = Integer.parseInt(playerInfo[2]); // number of games to play
    int[] allGame = new int[totalGame];
    boolean machineWon;
    String winner = null;
    for (int i = 1; i<=totalGame; i++)
        {
            System.out.println("Game " + i);
            do{ // so all variables are initialised
                leftChocolates = Machine(random, leftChocolates); // machines turn
                machineWon = true;
                if (leftChocolates<=0) // checks if there are any chocolates left
                {
                    break; // breaks do while loop
                }
                leftChocolates = Player(scanner, leftChocolates); // players turn
                machineWon = false;
            } while (!(leftChocolates<=0)); // runs until there is no more chocolates left
        leftChocolates = Integer.parseInt(playerInfo[1]); // gets original starting number of chocolates as new round
        allGame = Score(playerInfo,machineWon,allGame); //records who won
        winner = Result(playerInfo,allGame);
        }
    if (winner.equals("Draw")) // draw
    {
        System.out.println("Its a draw");
    }else
    {
        System.out.println("Winner is " + winner); // someone won
    }
    return;
} // END GameLoop
````
